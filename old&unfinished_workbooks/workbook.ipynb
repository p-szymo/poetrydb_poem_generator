```python
def poem_grabber(title):
    base_url = 'http://poetrydb.org/title/'
    url_addon = (title + '/lines.json').replace(" ", "%20")
    try:
        response = requests.get(base_url + url_addon)
        lines = response.json()[0]['lines']
        poem = " \n ".join(lines).lower()
        poem = poem.translate(str.maketrans('', '', string.punctuation)).replace('    ', '\t ').replace('  ', ' ').replace(' —', '')
    except:
        print(title)
        pass
    return poem
```
